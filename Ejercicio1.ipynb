## Pregunta 4

Una de las principales labores de la Unidad Nacional para la GestiÃ³n del Riesgo de Desastres (UNGRD), es prestar ayuda econÃ³mica para satisfacer diferentes necesidades al momento de presentar eventos.

Â¿QuÃ© porcentaje del total de los recursos ejecutados por la UNGRD en el aÃ±o 2019 corresponde a ayudas asociadas a kits de alimentos y quÃ© porcentajes a materiales de construcciÃ³n?

```python
# funcion para arreglar las columnas 
def convert_values(x):
    if type(x) == int:
        return x
    else:
        x = x.split()[-1]
        if x == "0":
            return int(x)
        else:
            return int(x.split()[-1].replace(",","")[:-3])
    
```

```python
data_2019["RECURSOS EJECUTADOS"] = data_2019["RECURSOS EJECUTADOS"].apply(convert_values)
data_2019["VALOR KIT DE ALIMENTO"] = data_2019["VALOR KIT DE ALIMENTO"].apply(convert_values)
data_2019["VALOR MATERIALES DE CONSTRUCCION"] = data_2019["VALOR MATERIALES DE CONSTRUCCION"].apply(convert_values)
data_2019
```

```python
recursos_2019 = data_2019["RECURSOS EJECUTADOS"].sum()
kits_alimentos_2019 = data_2019["VALOR KIT DE ALIMENTO"].sum()
materiales_constr_2019 = data_2019["VALOR MATERIALES DE CONSTRUCCION"].sum()

print(f"Recursos 2019 = {recursos_2019:,.0f}")
print(f"Kits 2019 = {kits_alimentos_2019:,.0f}")
print(f"Construccion 2019 = {materiales_constr_2019:,.0f}")

# data_2019[data_2019["VALOR KIT DE ALIMENTO"] == "$    35,100,000.00"]
```

```python
p_kits = kits_alimentos_2019/recursos_2019 * 100
print(f"Porcentaje de kits de alimentos = {p_kits:.1f} %")

p_materiales = materiales_constr_2019/recursos_2019 * 100000
print(f"Porcentaje de materiales = {p_materiales:.1f} %")

# Debe refinarse la funciÃ³n para que los materiales de construcciÃ³n den bien
# falla por tres ordenes de magnitud que es el [:-3] de la funcion, no aplica a esta columna
```

## Pregunta 9

Parece que el funcionamiento de la UNGRD no solo cometiÃ³ el olvido del rÃ³tulo adecuado en los anteriores grÃ¡ficos de barras sino que tambiÃ©n lo hizo con este grÃ¡fico de torta

Para responder a esta pregunta se hace descartando y graficando cual de estos estÃ¡ bien. Se empieza por la primera opciÃ³n

```python
tol_eventos = len(data["EVENTO"])
tol_eventos_2019 = len(data["EVENTO"][data["AÃ‘O"]==2019])
tol_eventos_2020 = len(data["EVENTO"][data["AÃ‘O"]==2020])
tol_eventos_2021 = len(data["EVENTO"][data["AÃ‘O"]==2021])

p_2019 = tol_eventos_2019/tol_eventos * 100
p_2020 = tol_eventos_2020/tol_eventos * 100
p_2021 = tol_eventos_2021/tol_eventos * 100

print(f"Porcentaje eventos 2019 = {p_2019:.2f}")
print(f"Porcentaje eventos 2020 = {p_2020:.2f}")
print(f"Porcentaje eventos 2021 = {p_2021:.2f}")
```

Parece que no es la primera, se verifica la segunda

```python
tol_eventos = data["FALLECIDOS"].sum()
tol_eventos_2019 = data["FALLECIDOS"][data["AÃ‘O"]==2019].sum()
tol_eventos_2020 = data["FALLECIDOS"][data["AÃ‘O"]==2020].sum()
tol_eventos_2021 = data["FALLECIDOS"][data["AÃ‘O"]==2021].sum()

p_2019 = tol_eventos_2019/tol_eventos * 100
p_2020 = tol_eventos_2020/tol_eventos * 100
p_2021 = tol_eventos_2021/tol_eventos * 100

print(f"Porcentaje eventos 2019 = {p_2019:.2f}")
print(f"Porcentaje eventos 2020 = {p_2020:.2f}")
print(f"Porcentaje eventos 2021 = {p_2021:.2f}")
```
